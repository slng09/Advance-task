```python
import pandas as pd
import numpy as np
import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt
import seaborn as sns

#--- suppress warnings ---
import warnings
warnings.filterwarnings("ignore")
```

```python
# Read CSV
from google.colab import files
uploaded = files.upload()
import io
df = pd.read_csv(io.BytesIO(uploaded["TA_M_B.csv"]),sep=',')
df.head()
```

```python
#rename column names
df.columns = ['stop_id', 'stop_name', 'latitude', 'longtitude', 'mode', 'state_code'] 
df.columns
```

```python
#drop missing value
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
def clean_column_name(column_names):
    clean_column_names = []
    for c in column_names:
        c = c.lower().replace(' ','_')
        clean_column_names.append(c)
    return clean_column_names    

df.columns = clean_column_name(df.columns)
```

```python
df.info()
```

```python
duplicate = df[df.duplicated()]
print("Duplicate Rows :")
duplicate
```

```python
#--- take the means to replace the zero ---

#df['stop_name'] = df['stop_name'].replace([np.nan], )
df['mode'] = df['mode'].replace([np.nan], ['train'] )
df['state_code'] = df['state_code'].replace([np.nan], ['TA'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Replace NaN with zero on all columns 
df = df.fillna(0)
```
